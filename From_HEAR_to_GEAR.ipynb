# GEAR (Generative Evaluation of Audio Representations)

Define the embedding calculator

```python
from hearbaseline import wav2vec2
import torch
import hearbaseline
from GURA import fusion_cat_xwc
import serab_byols
from hearbaseline import torchcrepe
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import openl3

def cal_embeddings(model_name, X):
    X = torch.from_numpy(X).float()
    if model_name == "hearbaseline":
        model = hearbaseline.load_model(
                "hear-baseline/saved_models/naive_baseline.pt")
        embeddings = hearbaseline.get_scene_embeddings(X, model)
    
    elif model_name == "openl3":
        embeddings, _ = openl3.get_audio_embedding(list(X.numpy()), 16000)
        # average at time dimension axis=1
        embeddings = np.mean(np.array(embeddings), axis=1)
        embeddings = np.array(embeddings)
        model = None

    elif model_name == "GURA":
        model = fusion_cat_xwc.load_model()
        embeddings = fusion_cat_xwc.get_scene_embeddings(X, model)

    elif model_name == "byol":
        X = X.cuda()
        model_name = 'cvt'
        checkpoint_path = "serab-byols/checkpoints/cvt_s1-d1-e64_s2-d1-e256_s3"\
        "-d1-e512_BYOLAs64x96-osandbyolaloss6373-e100-bs256-lr0003-rs42.pth"
        model = serab_byols.load_model(checkpoint_path, model_name)
        embeddings = serab_byols.get_scene_embeddings(X, model)
        embeddings = embeddings.cpu()

    elif model_name == "crepe":
        model = torchcrepe.load_model()
        embeddings = torchcrepe.get_scene_embeddings(X, model)

    elif model_name == "wav2vec2":
        X = X.cuda()
        model = wav2vec2.load_model()
        embeddings = wav2vec2.get_scene_embeddings(X, model)
        embeddings = embeddings.cpu()

    elif model_name == "yamnet":
        # referencing tensorflow document: 
        # https://www.tensorflow.org/hub/tutorials/yamnet
        embeddings = []
        model = hub.load('https://tfhub.dev/google/yamnet/1')
        # using MaxAbsScalar to make sure the range of each feature is in
        # [-1, 1], which is the type of data required by yamnet model
        # defined in the document
        for i in range(X.shape[0]):
            scores, embedding, spectrogram = model(X[i])
            embeddings.append(embedding[0])
        embeddings = np.array(embeddings)

    del model
    return embeddings
```

Ensure range of values are in [-1, 1]

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
X_standarlized = scaler.fit_transform(X)
```

Calculating embeddings

```python
base_embedding = cal_embeddings("hearbaseline", X_standarlized)
openl3_embedding = cal_embeddings("openl3", X_standarlized)
gura_embedding = cal_embeddings("GURA", X_standarlized)
byol_embedding = cal_embeddings("byol", X_standarlized)
crepe_embedding = cal_embeddings("crepe", X_standarlized)
wav2vec2_embedding = cal_embeddings("wav2vec2", X_standarlized)
yamnet_embedding = cal_embeddings("yamnet", X_standarlized)

print("base embeddings shape {}".format(base_embedding.shape))
print("openl3 embeddings shape {}".format(openl3_embedding.shape))
print("gura embeddings shape {}".format(gura_embedding.shape))
print("byol embeddings shape {}".format(byol_embedding.shape))
print("crepe embeddings shape {}".format(crepe_embedding.shape))
print("wav2vec2 embeddings shape {}".format(wav2vec2_embedding.shape))
print("yamnet embeddings shape {}".format(yamnet_embedding.shape))
```

```python
embeddings = {"base": base_embedding, 
              r"Open-$L^3$": openl3_embedding, # here raw text for mathimatical representation in plots
              "BYOL": byol_embedding, "CREPE": crepe_embedding,
              "Wav2vec2": wav2vec2_embedding, 
              "GURA": gura_embedding, 
              "YAMNet": yamnet_embedding}
```

Quick visualization of dimension-reduced embeddings
